```julia
n = 10

A = round.(randn(n,n)*10)
```

```julia
using LinearAlgebra
using Statistics
det(A)
```

```julia
A[2,:] = A[1,:]*5;

det(A)
```

```julia
rank(A)
```

```julia
lam = Float64[]
dets = Float64[]

for l = 0:.01:.1
    dets = Float64[]
    for k in 1:10
        shift = I(n)*l
        As = A + shift
        push!(dets, det(As))
    end
    push!(lam, mean(dets))
end
```
