```julia
using PlotlyJS

density = 50

S = [-7 12;12 -9];

range = -2:1/density:2
sz = length(range)
qf_mat = zeros(sz,sz)

# Compute quadratic form for w vectors between [-2 -2] and [2 2] 
for (ii, w1) in enumerate(range)
    for (jj, w2) in enumerate(range)
        w = [w1,w2]
        qf = (w'S*w) / w'w
        qf_mat[ii,jj] = qf
    end
end

# Create a surface plot
surface_plot = surface(z=qf_mat)


PlotlyJS.plot(surface_plot)
```
