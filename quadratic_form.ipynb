```julia
using PlotlyJS

density = 20

# Sample data for surface plot
x = 1:1:density 
y = 1:1:density 
z = rand(density, density)


# Set Z-axis limits
layout = Layout(
    scene = attr(
        zaxis = attr(range = [-2, 2])  # Setting the Z-axis limits
    )
)

# Create a surface plot
surface_plot = surface(x=x, y=y, z=z)


PlotlyJS.plot(surface_plot, layout)
```
