```julia
using LinearAlgebra
using Plots
```

```julia
m = 5
A = Symmetric(rand(m,m))
```

```julia
U, E, Vt = svd(A)
```

```julia
U = abs.(U)
E = abs.(E)
Vt = abs.(Vt)
```

```julia
L, W = eigen(A)
sorted_indices = sortperm(abs.(L), rev=true)
W = W[:,sorted_indices]
L = abs.(L[sorted_indices])
```

```julia
W = abs.(W)
```

```julia
U
```

```julia
Vt
```

```julia
plotlyjs()

U_plt = heatmap(abs.(U), title="U", color=:viridis)
E_plt = heatmap(Diagonal(E), title="E", color=:viridis)
Vt_plt = heatmap(abs.(Vt), title="Vt", color=:viridis)

L_plt = heatmap(Diagonal(L), title="L", color=:viridis)
W_plt = heatmap(W, title="W", color=:viridis)
UVt_plt = heatmap(U*Vt, title="U*E*Vt", color=:viridis)

A_plt = heatmap(U*Diagonal(E)*Vt', title="U*E*Vt'", color=:viridis)
A_plt2 = heatmap(W*Diagonal(L)*W', title="W*L*W'", color=:viridis)

# Combine subplots into a 2x3 layout
p = plot(U_plt, E_plt, Vt_plt, A_plt, L_plt, W_plt, A_plt2, layout=(2, 4), legend=false)

# Make plot square
plot!(p, size=(1400, 600))  # Set both width and height to the same value

# Display the combined plot
display(p)
```
